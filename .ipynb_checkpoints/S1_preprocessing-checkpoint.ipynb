```python
import mysql.connector
import nltk
import numpy as np
import pandas as pd
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = nltk.corpus.stopwords.words('french')
others = ['@', ':', 'de', 'ou']
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  passwd="password",
  database="twitter_db"
)

print(mydb)
```

```python
tweet_list = []

cursor = mydb.cursor()

query = ("SELECT text FROM tweets WHERE lang='fr' LIMIT 100")

cursor.execute(query)

for text in cursor:
    tweet_list.append(text[0])

cursor.close()
```

```python
tweets_tokens = [None] * 100
```

```python
for i in range(100):
    tweets_tokens[i] = nltk.word_tokenize(tweet_list[i])
```
